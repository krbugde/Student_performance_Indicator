# ✅ 1) What does `-e .` really mean?

When you run:

```
pip install -e .
```

Break it into pieces:

| Part          | Meaning           |
| ------------- | ----------------- |
| `pip install` | install something |
| `-e`          | editable mode     |
| `.`           | current folder    |

---

## 🧠 Editable Mode = Development Mode

Normally:

```
pip install myproject
```

copies code into site-packages.

But with:

```
pip install -e .
```

👉 Python **links** your project folder instead of copying it.

So:

• You change code in VS Code
• Immediately reflected when you run Python
• No need to reinstall each time

🔥 Very useful while developing ML projects.

---

## 📦 What exactly is installed?

Your project becomes importable:

```python
from ML_PROJECT.pipeline.train import train_model
```

Anywhere in that environment.

# ✅ 2) What is `.egg-info` folder?

When Python installs something, it stores **metadata**.

Metadata = information about package:

* name
* version
* dependencies
* where files live

Old format = `.egg-info`
Newer format = `.dist-info`

---

## 📁 Example:

After install:

```
ML_PROJECT.egg-info/
│
├── PKG-INFO
├── dependency_links.txt
├── requires.txt
├── top_level.txt
```

---

## 🧠 Why it exists?

Because pip needs to know:

✔ what is installed
✔ version
✔ uninstall later
✔ dependency tracking

You usually **don’t touch it**.

Think of it as the **receipt** of installation 🧾.

---

### 1️⃣ Industry standard for environments

In professional ML/Software projects:

**✅ Each project almost always has its own virtual environment** (or container, like Docker).

**Why?**

1. **Isolation of dependencies**

   * ML projects often need very specific versions of libraries (e.g., `TensorFlow 2.12`, `scikit-learn 1.3`, `numpy 1.27`).
   * If two projects share the same environment, installing/upgrading a library for one project could **break the other**.

2. **Reproducibility**

   * You want anyone (team member, CI/CD pipeline, or deployment server) to run your project and get **exactly the same behavior**.
   * This is usually ensured by environment files (`requirements.txt`, `environment.yml`, `Pipfile`) tied to a separate environment.

3. **Ease of deployment**

   * Separate environments make it **easy to package and deploy** ML applications without conflicts.

---

### 2️⃣ Typical practices

* **Python virtual environments**: `venv` or `conda` env per project.
* **Environment files**:

  * `requirements.txt` (pip)
  * `environment.yml` (conda)
    These files are committed to the project repo to recreate the environment anywhere.
* **Containers (advanced)**: Docker is often used in real-world projects to fully isolate environments with OS-level consistency.

**Example industry scenario:**

* Project A: uses TensorFlow 2.12, Python 3.10
* Project B: uses PyTorch 2.2, Python 3.9
* If you shared the same environment → conflicts occur
* Solution: each project has its own environment or Docker container

---

### 3️⃣ Summary

| Aspect          | Practice                                              |
| --------------- | ----------------------------------------------------- |
| Isolation       | ✅ Separate env per project                            |
| Reproducibility | ✅ Required, use `requirements.txt` or `conda env.yml` |
| Deployment      | ✅ Env or container ensures smooth deployment          |
| Disk usage      | ⚠️ Slightly more, but negligible compared to benefits |

💡 **Rule of thumb in industry:**

> “One project = One environment”
